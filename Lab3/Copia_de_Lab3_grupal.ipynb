```python
target = "AGTGTGTAACATTAGGGAGGACTTGAA"
secuence = secuence.replace("\n","") #O(n) 
                                     #Puede ser evitado haciendo formato
                                     #apropiado en la entrada

def Encontrar(target):
  for i in range(len(secuence)):   #O(n)
                                   #realizamos un leido de toda la secuencia
    if secuence[i] != target[0]:      #O(1)
                                      #Si el elemento que estamos analizando no
                                      #es igual al primero de nuestro objetivo
                                      #seguimos
      continue                          #O(1)
    for j in range(len(target)):      #O(n)
                                      #por cada elemento en nuestro objetivo
      if secuence[i+j] != target[j]:    #O(1)
                                        #revisamos si el elemento correspondiente de
                                        #la secuencia es igual
        break                             #O(1)
                                          #si no lo es, salimos de del ciclo
      if (j == len(target)-1):          #O(1)
                                        #si llegamos al final sin encontrar
                                        #diferencias
        return(i)                         #O(1)
                                          #devolvemos el indice donde empezamos


j = Encontrar(target)  
print(f"j = {j}")

print(secuence[j:j+len(target)])

target1 = "TTTAGTAGTGCTATCCCCATGTGATTTTAATAGCTTCTTAGGAGAATGACAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA"

i = Encontrar(target1)
print(f"i = {i}")
print(secuence[i:i + len(target1)])

target2 = "CAGGAGATGCCACAACTGCTTATGCTAATAGTGTTTTTAACATTTGTCAAG"

i = Encontrar(target2)
print(f"i = {i}")
print(secuence[i:i + len(target2)])

target2 = "TAAAGGTTTATACCTTCC"
i = Encontrar(target2)
print(f"i = {i}")
print(secuence[i:i + len(target2)])
```
